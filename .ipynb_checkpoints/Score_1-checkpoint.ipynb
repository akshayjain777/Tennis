# Data transformation functions

```python
### function to generate key for data
def generate_key(data):    
    data["key"]=data["matchid"]+'/'+data["set"].astype(str)+'/'+data["game"].astype(str)+'/'+data["point"].astype(str)+'/'+data["serve"].astype(str)
    data.sort_values('key',inplace=True, ascending=True)
    return data
```

```python
### just get the serves for a rally
def arc_123(data): 
    data=data[data['matchid'].str.contains('MS')]
    data=data[data['arcid']<4]
    data=data[data['arcid']!=0]                               
    return data
```

```python
### transform the rows to columns for arc 2 and 3
def arctoserve(data): 
        data1=data[data['arcid']==1]
        data2=data[data['arcid']==2]
        data3=data[data['arcid']==3]
        relev=['starttime',
         'endtime',
         'arcsvamatrixxpos',
         'arcsvamatrixxvel',
         'arcsvamatrixxaccel',
         'arcsvamatrixxchangeaccel',
         'arcsvamatrixypos',
         'arcsvamatrixyvel',
         'arcsvamatrixyaccel',
         'arcsvamatrixychangeaccel',
         'arcsvamatrixzpos',
         'arcsvamatrixzvel',
         'arcsvamatrixzaccel',
         'arcsvamatrixzchangeaccel','arcid','key' ]
        data2=data2[relev]
        data3=data3[relev]
        data2.columns = data2.columns.values + '2'
        data3.columns = data3.columns.values + '3'
        
        data2.rename(columns = {'key2' : 'key'}, inplace = True)
        data3.rename(columns = {'key3' : 'key'}, inplace = True)
        
        data4=pd.merge(data1,data2,how='inner',on='key')
        data4=pd.merge(data4,data3,how='inner',on='key')
        return data4
```

```python
### add player trajectories for particular shots.
def combine(df_prj,df_trj):
    print(df_prj.shape)
    df_prj=df_prj[df_prj['matchid'].str.contains('MS')]   

    num_prj=list(df_prj['key'].unique())
    df_trj=df_trj[df_trj['key'].isin(num_prj)]
    print('serve data completed')
    num_trj=list(df_trj['key'].unique())

    df_prj=df_prj[df_prj['key'].isin(num_trj)]
    df_prj_temp=df_prj[['matchtime','player1coordinatex',
     'player1coordinatey',
     'player1coordinatez',
     'player2coordinatex',
     'player2coordinatey',
     'player2coordinatez','key']]


    def plr_arc1_strt(i):                  #Function to interpolate and extract player coordinates 
        try:
            sttime=df_serve_temp[df_serve_temp['key']==i]['starttime'].values[0]
            df_temp=df_prj_temp[df_prj_temp['key']==i]
            aval=df_temp['matchtime'][df_temp['matchtime']>sttime].min()
            bval=df_temp['matchtime'][df_temp['matchtime']<sttime].max()

            if((aval-bval)!=0):
                p=(sttime-bval)/(aval-bval)
            else:
                p=0

            if( (len(df_temp)!=0) & (aval!=0) & (bval!=0) ):
                X1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatex'].values[0]
                X1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatex'].values[0]
                Y1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatey'].values[0]
                Y1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatey'].values[0]
                Z1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatez'].values[0]
                Z1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatez'].values[0]
                X2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatex'].values[0]
                X2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatex'].values[0]
                Y2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatey'].values[0]
                Y2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatey'].values[0]
                Z2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatez'].values[0]
                Z2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatez'].values[0]
                X1=X1_b+p*(X1_a-X1_b)
                Y1=Y1_b+p*(Y1_a-Y1_b)
                Z1=Z1_b+p*(Z1_a-Z1_b)
                X2=X2_b+p*(X2_a-X2_b)
                Y2=Y2_b+p*(Y2_a-Y2_b)
                Z2=Z2_b+p*(Z2_a-Z2_b)
                a=[X1,Y1,Z1,X2,Y2,Z2]
                print('!',end="")
                return a
        except Exception:
            b=[0,0,0,0,0,0]
            return b


    def plr_arc2_strt(i):                  #Function to interpolate and extract player coordinates 

        try:
            sttime=df_serve_temp[df_serve_temp['key']==i]['starttime2'].values[0]
            df_temp=df_prj_temp[df_prj_temp['key']==i]
            aval=df_temp['matchtime'][df_temp['matchtime']>sttime].min()
            bval=df_temp['matchtime'][df_temp['matchtime']<sttime].max()

        
            if((aval-bval)!=0):
                p=(sttime-bval)/(aval-bval)
            else:
                p=0

        
            if( (len(df_temp)!=0) & (aval!=0) & (bval!=0) ):
                X1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatex'].values[0]
                X1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatex'].values[0]
                Y1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatey'].values[0]
                Y1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatey'].values[0]
                Z1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatez'].values[0]
                Z1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatez'].values[0]
                X2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatex'].values[0]
                X2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatex'].values[0]
                Y2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatey'].values[0]
                Y2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatey'].values[0]
                Z2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatez'].values[0]
                Z2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatez'].values[0]
                X1=X1_b+p*(X1_a-X1_b)
                Y1=Y1_b+p*(Y1_a-Y1_b)
                Z1=Z1_b+p*(Z1_a-Z1_b)
                X2=X2_b+p*(X2_a-X2_b)
                Y2=Y2_b+p*(Y2_a-Y2_b)
                Z2=Z2_b+p*(Z2_a-Z2_b)
                a=[X1,Y1,Z1,X2,Y2,Z2]
                print('!',end="")
                return a
        except Exception:
                 b=[0,0,0,0,0,0]
                 return b

    def plr_arc3_end(i):                  #Function to interpolate and extract player coordinates 

        sttime=df_serve_temp[df_serve_temp['key']==i]['endtime3'].values[0]
        df_temp=df_prj_temp[df_prj_temp['key']==i]
        aval=df_temp['matchtime'][df_temp['matchtime']>sttime].min()
        bval=df_temp['matchtime'][df_temp['matchtime']<sttime].max()

        if((aval-bval)!=0):
            p=(sttime-bval)/(aval-bval)
        else:
            p=0

        try:
            if( (len(df_temp)!=0) & (aval!=0) & (bval!=0) ):
                X1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatex'].values[0]
                X1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatex'].values[0]
                Y1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatey'].values[0]
                Y1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatey'].values[0]
                Z1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatez'].values[0]
                Z1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatez'].values[0]
                X2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatex'].values[0]
                X2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatex'].values[0]
                Y2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatey'].values[0]
                Y2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatey'].values[0]
                Z2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatez'].values[0]
                Z2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatez'].values[0]
                X1=X1_b+p*(X1_a-X1_b)
                Y1=Y1_b+p*(Y1_a-Y1_b)
                Z1=Z1_b+p*(Z1_a-Z1_b)
                X2=X2_b+p*(X2_a-X2_b)
                Y2=Y2_b+p*(Y2_a-Y2_b)
                Z2=Z2_b+p*(Z2_a-Z2_b)
                a=[X1,Y1,Z1,X2,Y2,Z2]
                print('!',end="")
                return a
        except Exception:
                 b=[0,0,0,0,0,0]
                 return b

    def plr_arc3_strt(i):                  #Function to interpolate and extract player coordinates 

        sttime=df_serve_temp[df_serve_temp['key']==i]['starttime3'].values[0]
        df_temp=df_prj_temp[df_prj_temp['key']==i]
        aval=df_temp['matchtime'][df_temp['matchtime']>sttime].min()
        bval=df_temp['matchtime'][df_temp['matchtime']<sttime].max()

        if((aval-bval)!=0):
            p=(sttime-bval)/(aval-bval)
        else:
            p=0

        try:
            if( (len(df_temp)!=0) & (aval!=0) & (bval!=0) ):
                X1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatex'].values[0]
                X1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatex'].values[0]
                Y1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatey'].values[0]
                Y1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatey'].values[0]
                Z1_a=df_temp[df_temp['matchtime']==aval]['player1coordinatez'].values[0]
                Z1_b=df_temp[df_temp['matchtime']==bval]['player1coordinatez'].values[0]
                X2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatex'].values[0]
                X2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatex'].values[0]
                Y2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatey'].values[0]
                Y2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatey'].values[0]
                Z2_a=df_temp[df_temp['matchtime']==aval]['player2coordinatez'].values[0]
                Z2_b=df_temp[df_temp['matchtime']==bval]['player2coordinatez'].values[0]
                X1=X1_b+p*(X1_a-X1_b)
                Y1=Y1_b+p*(Y1_a-Y1_b)
                Z1=Z1_b+p*(Z1_a-Z1_b)
                X2=X2_b+p*(X2_a-X2_b)
                Y2=Y2_b+p*(Y2_a-Y2_b)
                Z2=Z2_b+p*(Z2_a-Z2_b)
                a=[X1,Y1,Z1,X2,Y2,Z2]
                print('!',end="")
                return a
        except Exception:
                 b=[0,0,0,0,0,0]
                 return b

    df_serve_temp=df_trj[['starttime','key','endtime','starttime2','starttime3','endtime3']]

    df_serve_temp['Coordinates startarc1']=df_serve_temp['key'].apply(lambda x:plr_arc1_strt(x))
    print("done 1")

    df_serve_temp['Coordinates startarc2']=df_serve_temp['key'].apply(lambda x:plr_arc2_strt(x))
    print('done 2')

    df_serve_temp['Coordinates startarc3']=df_serve_temp['key'].apply(lambda x:plr_arc3_strt(x))
    print('done 3')

    df_serve_temp['Coordinates endarc3']=df_serve_temp['key'].apply(lambda x:plr_arc3_end(x))
    print("done 4")

    df_serve_temp=df_serve_temp.drop(['starttime','starttime2','starttime3','endtime','endtime3'],axis=1)

    df_serve_f = pd.merge(df_trj,df_serve_temp, how='inner', on=['key'])    

    return df_serve_f
```

# Features creation for shots scoring

## FEATURES

### Time, y_cord, z_cord

```python
def xy(data): ### FUNCTION TO CREATE TIME, Y AND Z CORD AT BASELINE
    a= []
    for i in range(0,len(data)):
        t = np.roots([data.arcsvamatrixxchangeaccel3[i],data.arcsvamatrixxaccel3[i],data.arcsvamatrixxvel3[i],(data.arcsvamatrixxpos3[i]-11.89)])
        real = t[np.isreal(t)]
        for j in range(0,len(real)):
            time = np.where([data['starttime3'][i] < real[j]], real[j],data['endtime3'][i] )
        a.append(time)    
    a = pd.DataFrame(a)    
    
    data['time'] = a
    data['time'] = data['time'].real        
    data['z_cord'] = (data['arcsvamatrixzpos3']+data['arcsvamatrixzvel3']*data['time']+data['arcsvamatrixzaccel3']*(data['time']**2)+data['arcsvamatrixzchangeaccel3']*data['time']**3)
    data['y_cord'] = (data['arcsvamatrixypos3']+data['arcsvamatrixyvel3']*data['time']+data['arcsvamatrixyaccel3']*(data['time']**2)+data['arcsvamatrixychangeaccel3']*data['time']**3)
    return data
```

```python
serves_data1['initialxposition']=serves_data1['arcsvamatrixxpos']+serves_data1['arcsvamatrixxvel']*serves_data1['starttime']+serves_data1['arcsvamatrixxaccel']*(serves_data1['starttime']**2)+serves_data1['arcsvamatrixxchangeaccel']*(serves_data1['starttime']**3)
serves_data1['initialyposition']=serves_data1['arcsvamatrixypos']+serves_data1['arcsvamatrixyvel']*serves_data1['starttime']+serves_data1['arcsvamatrixyaccel']*(serves_data1['starttime']**2)+serves_data1['arcsvamatrixychangeaccel']*(serves_data1['starttime']**3)
serves_data1['initialzposition']=serves_data1['arcsvamatrixzpos']+serves_data1['arcsvamatrixzvel']*serves_data1['starttime']+serves_data1['arcsvamatrixzaccel']*(serves_data1['starttime']**2)+serves_data1['arcsvamatrixzchangeaccel']*(serves_data1['starttime']**3)

serves_data1['initialxposition2']=serves_data1['arcsvamatrixxpos2']+serves_data1['arcsvamatrixxvel2']*serves_data1['starttime2']+serves_data1['arcsvamatrixxaccel2']*(serves_data1['starttime2']**2)+serves_data1['arcsvamatrixxchangeaccel2']*(serves_data1['starttime2']**3)
serves_data1['initialyposition2']=serves_data1['arcsvamatrixypos2']+serves_data1['arcsvamatrixyvel2']*serves_data1['starttime2']+serves_data1['arcsvamatrixyaccel2']*(serves_data1['starttime2']**2)+serves_data1['arcsvamatrixychangeaccel2']*(serves_data1['starttime2']**3)
serves_data1['initialzposition2']=serves_data1['arcsvamatrixzpos2']+serves_data1['arcsvamatrixzvel2']*serves_data1['starttime2']+serves_data1['arcsvamatrixzaccel2']*(serves_data1['starttime2']**2)+serves_data1['arcsvamatrixzchangeaccel2']*(serves_data1['starttime2']**3)

serves_data1['initialxposition3']=serves_data1['arcsvamatrixxpos3']+serves_data1['arcsvamatrixxvel3']*serves_data1['starttime3']+serves_data1['arcsvamatrixxaccel3']*(serves_data1['starttime3']**2)+serves_data1['arcsvamatrixxchangeaccel3']*(serves_data1['starttime3']**3)
serves_data1['initialyposition3']=serves_data1['arcsvamatrixypos3']+serves_data1['arcsvamatrixyvel3']*serves_data1['starttime3']+serves_data1['arcsvamatrixyaccel3']*(serves_data1['starttime3']**2)+serves_data1['arcsvamatrixychangeaccel3']*(serves_data1['starttime3']**3)
serves_data1['initialzposition3']=serves_data1['arcsvamatrixzpos3']+serves_data1['arcsvamatrixzvel3']*serves_data1['starttime3']+serves_data1['arcsvamatrixzaccel3']*(serves_data1['starttime3']**2)+serves_data1['arcsvamatrixzchangeaccel3']*(serves_data1['starttime3']**3)
```

```python
serves_data1['arc3_finalx']=serves_data1['arcsvamatrixxpos3']+serves_data1['arcsvamatrixxvel3']*serves_data1['endtime3']+serves_data1['arcsvamatrixxaccel3']*(serves_data1['endtime3']**2)+serves_data1['arcsvamatrixxchangeaccel3']*(serves_data1['endtime3']**3)
serves_data1['arc3_finaly']=serves_data1['arcsvamatrixypos3']+serves_data1['arcsvamatrixyvel3']*serves_data1['endtime3']+serves_data1['arcsvamatrixyaccel3']*(serves_data1['endtime3']**2)+serves_data1['arcsvamatrixychangeaccel3']*(serves_data1['endtime3']**3)
serves_data1['arc3_finalz']=serves_data1['arcsvamatrixzpos3']+serves_data1['arcsvamatrixzvel3']*serves_data1['endtime3']+serves_data1['arcsvamatrixzaccel3']*(serves_data1['endtime3']**2)+serves_data1['arcsvamatrixzchangeaccel3']*(serves_data1['endtime3']**3)
```

```python
serves_data1['m1']=(serves_data1['arc3_finaly']-serves_data1['initialyposition3'])/(serves_data1['arc3_finalx']-serves_data1['initialxposition3'])
serves_data1['m2']=(serves_data1['initialyposition2']-serves_data1['initialyposition'])/(serves_data1['initialxposition2']-serves_data1['initialxposition'])
serves_data1['Spin Angle']=np.arctan(np.abs(serves_data1['m2']-serves_data1['m1'])/(1+serves_data1['m2']*serves_data1['m1']))
serves_data1=serves_data1.drop(['m1','m2'],axis=1)
serves_data1['Spin Angle'] = np.degrees(serves_data1['Spin Angle'])
```

```python
### TO CALCULATE WHETHER THE BALL IS GOING TOWARDS THE PLAYER OR AWAY FROM THE PLAYER
for i in range(0,len(serves_data1)):
    if (serves_data1['initialyposition2'][i] >= 0):
        if ((serves_data1['initialyposition2'][i] < 2) & (serves_data1['initialyposition2'][i] > 0)) &(serves_data1['arc3_finaly'][i] - serves_data1['initialyposition2'][i] > 1):
            serves_data1['Spin Angle'][i] = serves_data1['Spin Angle'][i] * 0.5
        if ((serves_data1['initialyposition2'][i] > 3) & (-1 < serves_data1['arc3_finaly'][i] - serves_data1['initialyposition2'][i] < 1)):
            serves_data1['Spin Angle'][i] = serves_data1['Spin Angle'][i] * 0.5
            
    if (serves_data1['initialyposition2'][i] < 0):
        if ((serves_data1['initialyposition2'][i] > -2) & (serves_data1['initialyposition2'][i] < 0)) &(serves_data1['arc3_finaly'][i] - serves_data1['initialyposition2'][i] < -1):
            serves_data1['Spin Angle'][i] = serves_data1['Spin Angle'][i] * 0.5
        if ((serves_data1['initialyposition2'][i] < -3) & (-1 < serves_data1['arc3_finaly'][i] - serves_data1['initialyposition2'][i] < 1)):
            serves_data1['Spin Angle'][i] = serves_data1['Spin Angle'][i] * 0.5
```

```python
serves_data1['serve2'] = np.where(serves_data1['serve'] == 2,0,1) ## SECOND SERVE CRITERIA
```

```python
serves_data1['dist from baseline'] = np.abs(11.85) - np.abs(serves_data1['initialxposition2']) ## DISTANCE FROM BASELINE
```
